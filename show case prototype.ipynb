```python
dfo = pd.read_csv('output.csv',index_col=False)
dfo.head()
dfo.columns=['no','name','act']
```

```python
dfo.head(100)
```

```python
dfo=dfo.astype(str).apply(lambda x: x.str.lower())
dfo=dfo.astype(str).apply(lambda x: x.str.replace(" ",""))
dfo['act'][11]='art'
dfo['act'][93]='science'
dfo['act']=dfo['act'].replace(['art','bb','computer','games','publicspk','science'],['artandcraft','beatboxing','basicsofcomputers','gamesandpuzzles','publicspeaking','scienceexperiments'])
```

```python
finalAct=dfo['act'].unique()
finalActObj={}

finalAct
```

```python
for finAct in finalAct:
    actSize=len(dfo.loc[dfo['act']==finAct])
    finalActObj[finAct]={'name':finAct,'total':actSize,'students':[],'diff':-actSize}
```

```python
import random

for i in range(0,len(df)):
    df.loc[i]['act1']=df.loc[i]['act1']+'-'+str(round(random.uniform(1.5,5.5)+5,1))
    df.loc[i]['act2']=df.loc[i]['act2']+'-'+str(round(random.uniform(1.5,5.5)+3,1))
    df.loc[i]['act3']=df.loc[i]['act3']+'-'+str(round(random.uniform(1.5,5.5)+1,1))

pd.set_option('display.height', 500)
pd.set_option('display.max_rows', 500)
df
```

```python
from pprint import pprint 
maxScoreList=[]
scoresAddedList=[]
for i in range(0,len(df)):
    s1=float(df.loc[i]['act1'].split('-')[1])
    s2=float(df.loc[i]['act2'].split('-')[1])
    s3=float(df.loc[i]['act3'].split('-')[1])
    maxScoreList=[s1,s2,s3]
    maxScoreList=sorted(maxScoreList,reverse=True)
#     print(maxScoreList)
    maxScore=maxScoreList[0]
    for act in [df.loc[i]['act1'],df.loc[i]['act2'],df.loc[i]['act3']]:
        if float(act.split('-')[1])==maxScore:
            scoresAddedList.append({'name':df.loc[i]['name'],'act':act.split('-')[0]})

print(scoresAddedList)
print('\n\n\n\n\n')
scoresAddedList1=[]
for i in range(0,len(df)):
    s1=float(df.loc[i]['act1'].split('-')[1])
    s2=float(df.loc[i]['act2'].split('-')[1])
    s3=float(df.loc[i]['act3'].split('-')[1])
    maxScoreList=[s1,s2,s3]
    maxScoreList=sorted(maxScoreList,reverse=True)
    maxScore=maxScoreList[1]
    for act in [df.loc[i]['act1'],df.loc[i]['act2'],df.loc[i]['act3']]:
        if float(act.split('-')[1])==maxScore:
            scoresAddedList1.append({'name':df.loc[i]['name'],'act':act.split('-')[0]})
            
print(scoresAddedList1)            
print('\n\n\n\n\n')
scoresAddedList2=[]
for i in range(0,len(df)):
    s1=float(df.loc[i]['act1'].split('-')[1])
    s2=float(df.loc[i]['act2'].split('-')[1])
    s3=float(df.loc[i]['act3'].split('-')[1])
    maxScoreList=[s1,s2,s3]
    maxScoreList=sorted(maxScoreList,reverse=True)
    print('3rd pass '+str(maxScoreList)+' leastmax:'+str(maxScoreList[2]))
    maxScore=maxScoreList[2]
    for act in [df.loc[i]['act1'],df.loc[i]['act2'],df.loc[i]['act3']]:
        if float(act.split('-')[1])==maxScore:
            scoresAddedList2.append({'name':df.loc[i]['name'],'act':act.split('-')[0]})
print('\n\n\n\n\n')
print(scoresAddedList2)
```
